# Query energy_prod.energy.v_sessions via Databricks

**Before running:** fill in your credentials in the `.env` file:
- `DATABRICKS_HOST` — workspace URL, e.g. `https://<workspace>.cloud.databricks.com`
- `DATABRICKS_TOKEN` — personal access token
- `DATABRICKS_HTTP_PATH` — SQL warehouse HTTP path (find in Databricks UI: SQL Warehouses → your warehouse → Connection details → HTTP path)

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)

host = os.getenv("DATABRICKS_HOST")
token = os.getenv("DATABRICKS_TOKEN")
http_path = os.getenv("DATABRICKS_HTTP_PATH")

assert host, "Set DATABRICKS_HOST in .env"
assert token, "Set DATABRICKS_TOKEN in .env"
assert http_path, "Set DATABRICKS_HTTP_PATH in .env"

print(f"Host: {host}")
print(f"Token set: {bool(token)}")
print(f"HTTP path set: {bool(http_path)}")
```

```python
import ssl
import certifi
from databricks import sql as databricks_sql
import pandas as pd

# Point SSL at certifi's CA bundle (fixes macOS Python SSL errors)
os.environ["SSL_CERT_FILE"] = certifi.where()
os.environ["REQUESTS_CA_BUNDLE"] = certifi.where()

QUERY = "SELECT * FROM energy_prod.energy.v_sessions s LIMIT 10"

with databricks_sql.connect(
    server_hostname=host.replace("https://", "").strip("/"),
    http_path=http_path,
    access_token=token,
) as conn:
    cursor = conn.cursor()
    cursor.execute(QUERY)
    rows = cursor.fetchall()
    column_names = [desc[0] for desc in cursor.description]

df = pd.DataFrame(rows, columns=column_names)
display(df)
```
